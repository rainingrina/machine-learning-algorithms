```python
import pandas as pd
df = pd.read_csv("csv files/titanic.csv")
df.head()
```

```python
df.shape
```

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns', inplace=True)
df.head()
```

```python
target = df.Survived
inputs = df.drop('Survived', axis='columns')
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
inputs = pd.concat([inputs, dummies], axis='columns')
inputs.head(3)
```

```python
inputs.drop('Sex', axis='columns', inplace=True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
inputs.head(6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(inputs)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
import pandas as pd
df = pd.read_csv("csv files/spam.csv")
df.head()
```

```python
df.groupby('Category').describe()
```

```python
#we need to convert both category and message to numbers
df['spam'] = df['Category'].apply(lambda x: 1 if x =='spam' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2)
```

```python
#count vectorize -> represents words as numbers -> takes all unique words only
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = [
        "Congratulations! You were amazing.",
        'Upto 20 percent discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
y_train
```

```python
X_train.values
```

```python
emails_count
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer',  CountVectorizer()),
    ('nb', MultinomialNB())

])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.predict(emails)
```

```python
clf.score(X_train, y_train)
```
